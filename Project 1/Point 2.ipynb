# UniProt can also be accessed programmatically

    -directly, see: https://www.uniprot.org/help/programmatic_access 
    -via the python Bioservices package: https://bioservices.readthedocs.io/en/master/
    
Here below some examples using bioservices

```python
from bioservices.uniprot import UniProt
u = UniProt(verbose=False)
```

```python
import pyhgnc
import pandas
from bioservices import *
```

```python
h=HGNC()
file = open('4.1-ID_list.txt', mode = 'r', encoding = 'utf-8-sig')
lines = file.readlines()
file.close()
```

```python
for i in range(len(lines)):
    lines[i]=lines[i].replace('\n','')
```

```python
def get_gene_info(genes_list,hgnc_obj):
    df=pandas.DataFrame()
    l=[]
    for line in lines:
        result=h.fetch('symbol', line)
        result=result['response']
        result=result['docs']
        l.append(line)
        l.append(result[0]['entrez_id'])
        l.append(result[0]['name'])
        l.append(result[0]['uniprot_ids'][0])
        res = u.search(result[0]['uniprot_ids'][0], frmt="txt")
        r=res.split(' ')
        f=[]
        for i in range(len(r)):
            if r[i]=='FUNCTION:':
                for j in range(1,20):
                    if r[i+j]!='':
                        f.append(r[i+j].replace('\nCC',''))
                    if r[i+j]=='\n':
                        break
                    if r[i+j]=='':
                        continue
                break
        l.append(' '.join(f))
        l.append(result[0]['date_approved_reserved'])
        df=df.append([l[0:6]])
        l=[]
    return df
    
```

```python
df=get_gene_info(lines,h)
```

```python
df.columns=["Gene","Entrez ID","Name","Uniprot ID","FUNCTION","Approved Date"]
```
