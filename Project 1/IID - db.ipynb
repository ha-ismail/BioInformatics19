```python
from bs4 import BeautifulSoup
import mechanicalsoup as ms 
import pickle
import csv
```

```python
import pandas as pd
```

```python
def save_obj(obj, name):
    with open('/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name):
    with open('/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
gList = []
#function to read the genes from the txt file and store them in a gList
with open('9genes_list.txt') as file:
        line = csv.reader(file, delimiter='\n')
        #print(*line)
        genes = list(line)
        for g in genes:
            if str(*g) != '' and not g[0][0].isdigit() :  gList.append(g[0])
```

```python
def web_scrap_list(query):
    browser = ms.StatefulBrowser()
    url = "http://iid.ophid.utoronto.ca/iid/Search_By_Proteins/"
    browser.open(url)
    #print(browser.get_current_page())
    browser.get_current_page().find_all('textarea')
    sf=browser.select_form()
    #sf.print_summary()
    sf.set_textarea({"ids_area":query})   # query is the line
    browser.submit_selected()
    # sf["includeSource"] = " Source information: detection methods, PubMed IDs, reporting databases"
    Results = browser.get_current_page()
    ppi_interactions=[]
    
    try:
        results_findall = Results.findAll('table')[0].findAll('tr')[1:]
    except Exception:
        return ppi_interactions
        
    count=0
    for row in results_findall:
        cells = row.find_all("td")
        
        qID, queryUP, interUP, querySym, interSym, species, evidence_type  = cells
        qID = qID.get_text()
        queryUP = queryUP.get_text()
        interUP = interUP.get_text()
        querySym = querySym.get_text()
        interSym = interSym.get_text()
        
        ppi_interactions.append([querySym, interSym, queryUP, interUP])
    return ppi_interactions
    
```

```python
# dictionary with keys Gene and values list of all interactions
genes_interactions = {}  
lst = gList
for el in lst:
    interactions = web_scrap_list(el) 
    genes_interactions[el] = interactions
    
```

```python
def create_df(d):
    df = pd.DataFrame(columns = ["QueryID","Query Uniprot", "Interactor Uniprot","Query Symbol", "Interactor Symbol"])
    i=0
    for key in d.keys():
        for lst in d[key]:
            df.loc[i] = [key]+lst
            i+=1
    return df
```

```python
create_df(genes_interactions)
```

```python
def summary(data): #data = genes_interactions
    tot_num_genes = len(data.keys())
    tot_interactions = sum([len(lst) for lst in data.values()])
    intersec = list()
    a=0
    seed_genes_found = []
    # find the total number of seed genes found in IID database:
    # We have the list, we need to check
    for rows in data.values():
        for lst in rows:
            if lst[1] in gList:
                seed_genes_found.append(1)
 
    a = sum(seed_genes_found)
    b = tot_num_genes
    c = tot_interactions
    return a,b,c
```

```python
summary(genes_interactions)
```
